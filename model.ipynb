````python
cols = [
    'no_peserta', 
    'no_keluarga', 
    'bobot', 
    'id_kunjungan_fktp', 
    'tgl_dtg_kunjungan_fktp', 
    'tgl_plg_kunjungan_fktp', 
    'provinsi_fktp', 
    'kota_fktp', 
    'kepemilikan_fktp', 
    'jenis_fktp', 
    'tipe_fktp', 
    'tingkat_pelayanan_fktp', 
    'jenis_poli_fktp', 
    'segmen_peserta_akses_layanan_fktp', 
    'kode_dan_nama_diagnosis', 
    'kode_diagnosis_tiga_digit', 
    'kode_diagnosis_tiga_sampai_lima_digit', 
    'nama_diagnosis', 
    'provinsi_faskes_rujukan', 
    'kota_faskes_rujukan', 
    'kepemilikan_faskes_rujukan', 
    'jenis_faskes_rujukan', 
    'tipe_faskes_rujukan', 
    'poli_faskes_rujukan', 
    'jenis_kunjungan_fktp', 
    'status_pulang',
]
````

````python
data = 'train_fktp.txt'
sep = ','
header = None

df = pd.read_csv(data, sep=sep, header=header, names=cols)
df.head()
````

````python
label_enc = LabelEncoder()

df.status_pulang = label_enc.fit_transform(df.status_pulang)
df.head()
````

````python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]

x.head()
````

````python
random_state = 104
test_size = 0.3
shuffle = True

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                    random_state=random_state,
                                    test_size=test_size,
                                    shuffle=shuffle)
````
